```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import datetime
```

```python
master = pd.read_csv('datasets/combined_lyft_dataset.csv', converters={'driver_onboard_date':pd.to_datetime, 'requested_at':pd.to_datetime, 'arrived_at':pd.to_datetime, 'picked_up_at':pd.to_datetime, 'dropped_off_at':pd.to_datetime})
master.head()
```

```python
keys = [
    'driver_id', # 0
    'ride_distance (miles)', # 1
    'ride_duration (minutes)', # 2
    'cost w/ prime time', # 3
    'lyft profit', # 4
    'driver_onboard_date', # 5
    'duration_request_to_accept', # 6
    'duration_accept_to_arrive', # 7
    'duration_arrived_to_pickup', # 8
    'duration_ride', # 9
    'worktime_expectancy (days)', # 10
    'number drives', # 11
    'group', # 12
    'dropped_off_at' # 13
    ]
```

```python
ma_unq_dids = master.groupby('driver_id').agg({
    keys[1] : 'mean',
    keys[2] : 'mean',
    keys[3] : 'sum',
    keys[4] : 'sum',
    keys[5] : 'max',
    keys[6] : 'mean',
    keys[7] : 'mean',
    keys[8] : 'mean',
    keys[9] : 'mean',
    keys[10] : 'mean',
    keys[11] : 'mean',
    keys[12] : 'max',
    keys[13] : 'max'
})
ma_unq_dids['worktime_expectancy (datetime)'] = ma_unq_dids['dropped_off_at'] - ma_unq_dids['driver_onboard_date']
ma_unq_dids.head()
```

# Find Targets

```python
a = 0
b = 0
c = 0
d = 0

def make_graphs(xdata, x):
    make_mo_graph(xdata, x)
    make_lt_graph(xdata, x)
    make_fq_graph(xdata, x)
    make_dr_graph(xdata, x)

def make_mo_graph(xdata, x):
    global a
    plt.figure(figsize=(10, 10))
    
    plt.scatter(xdata, ma_unq_dids['lyft profit'], marker='o', s=8, c=['r' if g == 0 else 'b' for g in ma_unq_dids['group']])
    plt.ylabel("Lyft's Profit (Dollars)")
    plt.xlabel(x)
    plt.title(x + " vs. Lyft's Profit (Dollars)")
    
    blue_patch = mpatches.Patch(color='blue', label='Drivers With 100 Or More Rides')
    red_patch = mpatches.Patch(color='red', label='Drivers With Less Than 100 Rides')
    black_patch = mpatches.Patch(color='black', label='Average Driver Value')
    plt.legend(handles=[blue_patch, red_patch, black_patch])
   
    z0 = np.polyfit(xdata.loc[ma_unq_dids['group'] == 0], ma_unq_dids['lyft profit'].loc[ma_unq_dids['group'] == 0], 1)
    p0 = np.poly1d(z0)
    plt.plot(xdata,p0(xdata),"r--")

    z = np.polyfit(xdata.loc[ma_unq_dids['group'] == 1], ma_unq_dids['lyft profit'].loc[ma_unq_dids['group'] == 1], 1)
    p = np.poly1d(z)
    plt.plot(xdata,p(xdata),"b--")
    
    plt.plot([min(xdata), max(xdata)], 637.04*np.ones(2), c='black')
    
    plt.savefig('a' + str(a) + '.png')
    a += 1
    plt.show()
    
def make_lt_graph(xdata, x):
    global b
    plt.figure(figsize=(10, 10))
    
    plt.scatter(xdata, ma_unq_dids[keys[10]], marker='o', s=8, c=['r' if g == 0 else 'b' for g in ma_unq_dids['group']])
    plt.ylabel("Driver Lifetime Expectancy (Days)")
    plt.xlabel(x)
    plt.title(x + " vs. Driver Lifetime Expectancy (Days)")
    
    blue_patch = mpatches.Patch(color='blue', label='Drivers With 100 Or More Rides')
    red_patch = mpatches.Patch(color='red', label='Drivers With Less Than 100 Rides')
    plt.legend(handles=[blue_patch, red_patch])
    
    z0 = np.polyfit(xdata.loc[ma_unq_dids['group'] == 0], ma_unq_dids[keys[10]].loc[ma_unq_dids['group'] == 0], 1)
    p0 = np.poly1d(z0)
    plt.plot(xdata,p0(xdata),"r--")
    
    z = np.polyfit(xdata.loc[ma_unq_dids['group'] == 1], ma_unq_dids[keys[10]].loc[ma_unq_dids['group'] == 1], 1)
    p = np.poly1d(z)
    plt.plot(xdata,p(xdata),"b--")
    
    plt.savefig('b' + str(b) + '.png')
    b += 1
    plt.show()

def make_fq_graph(xdata, x):
    global c
    plt.figure(figsize=(10, 10))
    
    plt.scatter(xdata, ma_unq_dids[keys[11]], marker='o', s=8, c=['r' if g == 0 else 'b' for g in ma_unq_dids['group']])
    plt.ylabel("Number of Drives")
    plt.xlabel(x)
    plt.title(x + " vs. Number of Drives")
    
    blue_patch = mpatches.Patch(color='blue', label='Drivers With 100 Or More Rides')
    red_patch = mpatches.Patch(color='red', label='Drivers With Less Than 100 Rides')
    plt.legend(handles=[blue_patch, red_patch])
    
    z0 = np.polyfit(xdata.loc[ma_unq_dids['group'] == 0], ma_unq_dids[keys[11]].loc[ma_unq_dids['group'] == 0], 1)
    p0 = np.poly1d(z0)
    plt.plot(xdata,p0(xdata),"r--")
    
    z = np.polyfit(xdata.loc[ma_unq_dids['group'] == 1], ma_unq_dids[keys[11]].loc[ma_unq_dids['group'] == 1], 1)
    p = np.poly1d(z)
    plt.plot(xdata,p(xdata),"b--")
    
    plt.savefig('c' + str(c) + '.png')
    c += 1
    plt.show()

def make_dr_graph(xdata, x):
    global d
    plt.figure(figsize=(10, 10))
    
    plt.scatter(xdata, ma_unq_dids[keys[6]] + ma_unq_dids[keys[7]], marker='o', s=8, c=['r' if g == 0 else 'b' for g in ma_unq_dids['group']])
    plt.ylabel("Time Between Drive Requested and Picked Up (Seconds)")
    plt.xlabel(x)
    plt.title(x + " vs. Time Between Drive Requested and Picked Up (Seconds)")
    
    blue_patch = mpatches.Patch(color='blue', label='Drivers With 100 Or More Rides')
    red_patch = mpatches.Patch(color='red', label='Drivers With Less Than 100 Rides')
    black_patch = mpatches.Patch(color='black', label='Average Wait Time')
    plt.legend(handles=[blue_patch, red_patch, black_patch])
    
    z0 = np.polyfit(xdata.loc[ma_unq_dids['group'] == 0], (ma_unq_dids[keys[6]] + ma_unq_dids[keys[7]]).loc[ma_unq_dids['group'] == 0], 1)
    p0 = np.poly1d(z0)
    plt.plot(xdata,p0(xdata),"r--")
    
    z = np.polyfit(xdata.loc[ma_unq_dids['group'] == 1],(ma_unq_dids[keys[6]] + ma_unq_dids[keys[7]]).loc[ma_unq_dids['group'] == 1], 1)
    p = np.poly1d(z)
    plt.plot(xdata,p(xdata),"b--")
    
    plt.plot([min(xdata), max(xdata)], 235.45*np.ones(2), c='black')
    
    plt.savefig('d' + str(d) + '.png')
    d += 1
    plt.show()
```

# Find Driver Lifetime Profile

```python
# Graph, for the average driver, the percentge of the total profit they made at each point in their lifetime in 10% intervals

def profile(var):
    ydata = np.zeros(11)
    for did in ma_unq_dids.index:
        lt = ma_unq_dids['worktime_expectancy (datetime)'].at[did]
        driver = master.loc[master['driver_id'] == did]
        var_sum = 0
        to_add = np.zeros(11)
        for i in range(len(driver)):
            ride = driver.iloc[i]
            percent_in = (ride['dropped_off_at'] - ride['driver_onboard_date']).total_seconds() / lt.total_seconds()
            to_add[int(percent_in // 10)] += ride[var]
            var_sum += ride[var]
        to_add /= var_sum
        ydata += to_add
    ydata /= len(ma_unq_dids.index)
    return ydata

profile('lyft profit')
```
